```python
from flask import Flask, jsonify
import pandas as pd
```

```python
app = Flask('__Case_API__')
```

```python
def load_data():
    return pd.read_csv('Salaries.csv')
```

```python
def calculate_salary_increase(data):
    data_2011 = data[data['Year'] == 2011]
    data_2014 = data[data['Year'] == 2014]
    total_pay_2011 = data_2011['TotalPay'].sum()
    total_pay_2014 = data_2014['TotalPay'].sum()
    return total_pay_2014 - total_pay_2011
```

```python
def highest_mean_salary_job(data):
    mean_salary_by_job = data.groupby('JobTitle')['TotalPay'].mean()
    return mean_salary_by_job.idxmax(), mean_salary_by_job.max()
```

```python
def calculate_overtime_savings_2014(data):
    data_2014 = data[data['Year'] == 2014]
    return data_2014['OvertimePay'].sum()
```

```python
def top_earning_employees(data):
    total_pay_by_employee = data.groupby('EmployeeName').agg({
        'JobTitle': 'first',
        'TotalPay': 'sum',
        'Year': 'first'
    }).sort_values(by='TotalPay', ascending=False).head(10)
    return total_pay_by_employee.reset_index()
```

```python
def least_common_jobs_cost_2014(data):
    data_2014 = data[data['Year'] == 2014]
    job_counts = data_2014['JobTitle'].value_counts().nsmallest(5)
    least_common_jobs = job_counts.index
    total_cost = data_2014[data_2014['JobTitle'].isin(least_common_jobs)].groupby('JobTitle')['TotalPay'].sum()
    return total_cost
```

```python
def overtime_pay_percentage(data):
    total_overtime_pay = data['OvertimePay'].sum()
    total_totalpay = data['TotalPay'].sum()
    return (total_overtime_pay / total_totalpay) * 100
```

```python
def results():
    # Load and preprocess data
    data = load_data()
    
    # Perform calculations
    salary_increase = calculate_salary_increase(data)
    highest_mean_salary_job_title, highest_mean_salary_value = highest_mean_salary_job(data)
    overtime_savings_2014 = calculate_overtime_savings_2014(data)
    top_earners = top_earning_employees(data)
    least_common_jobs_cost = least_common_jobs_cost_2014(data)
    overtime_percentage = overtime_pay_percentage(data)

    return render_template('results.html',
                           salary_increase=salary_increase,
                           highest_mean_salary_job_title=highest_mean_salary_job_title,
                           highest_mean_salary_value=highest_mean_salary_value,
                           overtime_savings_2014=overtime_savings_2014,
                           top_earners=top_earners,
                           least_common_jobs_cost=least_common_jobs_cost,
                           overtime_percentage=overtime_percentage)
```
