```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
dataset = pd.read_csv("Salaries.csv")
```

```python
dataset[["BasePay","Benefits"]] = dataset[["BasePay","Benefits"]].fillna(0)
```

```python
column_list = ["BasePay","OvertimePay","OtherPay","Benefits","TotalPay"]
```

```python
dataset=dataset[~np.all(dataset[['BasePay','OvertimePay','OtherPay','Benefits','TotalPay','TotalPayBenefits']]==0,axis=1)]
```

```python
dataset
```

```python
dataset = dataset[
    (dataset[column_list] >= 0).all(axis=1)
]
```

```python
dataset.isnull().sum() / len(dataset)
```

```python
total_salary_by_year = dataset.groupby('Year')['TotalPay'].sum().reset_index()
```

```python
total_salary_by_year
```

```python

plt.title('combined total salaries by year')
plt.xlabel('Year')
plt.ylabel('Total Salaries (billions)')

plt.plot(total_salary_by_year.Year, total_salary_by_year.TotalPay)
```

```python
print (total_salary_by_year['TotalPay'][3] - total_salary_by_year['TotalPay'][0])
print (((total_salary_by_year['TotalPay'][3] - total_salary_by_year['TotalPay'][0]) / total_salary_by_year['TotalPay'][3]) * 100)
```

```python
df_filtered = dataset[(dataset['Year'] >= 2011) & (dataset['Year'] <= 2014)]
mean_salaries = df_filtered.groupby('JobTitle')['TotalPay'].mean()
```

```python
highest_mean_salary_job = mean_salaries.idxmax()
highest_mean_salary = mean_salaries.max()
```

```python
print(f"The job with the highest mean salary from 2011 to 2014 is {highest_mean_salary_job} with a mean salary of {highest_mean_salary:.2f}")
```
